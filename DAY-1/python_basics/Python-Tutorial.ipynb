```python
# IF-ELSE

age = int(input('Enter your age: '))

if age >= 18:
  print('you can vote.')
elif age <= 0:
  print("you've not been born yet.")
else:
  print('you cannot vote.')


```

```python
# FOR-LOOP

for i in range(1,10+1):
  print(i)


name = input('Enter a fruit: ')
for j in name:
  print(j)
```

```python
# WHILE-LOOP
multiplier  = 5
number = 1 # loop variable
while number <= 10:
  print(number,'x',multiplier,'=',number * multiplier)
  number = number + 1

print('done')
```
